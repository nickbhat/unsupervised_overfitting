# API demo for Point Cloud Utilities 

Just a quick demo of how to use the point cloud utilities I wrote up. There are currently three functions

- `align`: SVD-based method for aligning two point clouds
- `rescale`: rescaling point clouds to have same range
- `dist`: Earth-mover distance between point clouds

Main features to add: 

- `get_null` functionality once we decide how to store simulations
- simulation API (Gaussians mixtures or whatever else we want)

I'll do all my demonstrations here on a simple mixture of two Gaussians in 2d. 

```python
import ot 
import numpy as np 
import matplotlib.pyplot as plt 
import pointcloud_utils as pcu
from sklearn.manifold import TSNE as tsne 
```

```python
# Generate some Gaussian mixture data 

# Number samples 
n = 100

# Parameters
means = np.array([[4.0, 4.0], [0.0, 0.0]])
covs = np.array([[[1.0, -0.8], [-.8, 1.0]], [[1.0, 0.0], [0.0, 1.0]]])

# Mixture proportions 
weights = np.array([0.5,0.5])

# Generate indices 
mixture_idx = np.random.choice(2, size = n, replace = True, p = weights)

# Number of samples from each component 
n_1 = sum(mixture_idx)
n_0 = n - n_1

# Generate 0,1 labels for downstream analysis
cluster_id = np.ones(n)
cluster_id[0:n_0] = 0

# Now sample 
x_0, y_0 = np.random.multivariate_normal(means[0], covs[0], n_0).T
x_1, y_1 = np.random.multivariate_normal(means[1], covs[1], n_1).T

# Generate arrays for each cluster 
c_0 = np.column_stack((x_0, y_0))
c_1 = np.column_stack((x_1, y_1))

# Massage into one data matrix, respecting the label order from above 
X = np.vstack((c_0, c_1))
```

```python
plt.plot(x_0, y_0, 'rx', x_1, y_1, 'bo')
plt.title("Original Data")
```

## Rescaling and Aligning. 

If we just run t-SNE on this data, we can get lots of geometry shift. Here is an example of two t-SNE's plotted together. 

```python
# Generate two t-SNE clusterings 
X_embed = tsne(n_components=2).fit_transform(X)
X_embed_2 = tsne(n_components=2).fit_transform(X)

# Plot them

fig, ax = plt.subplots() 

x = X_embed[:,0]
y = X_embed[:,1]
ax.scatter(x[cluster_id == 0], y[cluster_id == 0], marker = '^', color = 'red')
ax.scatter(x[cluster_id == 1], y[cluster_id == 1], marker = '+', color = 'red', label = 'tsne 1')

x = X_embed_2[:,0]
y = X_embed_2[:,1]
ax.scatter(x[cluster_id == 0], y[cluster_id == 0], marker = '^', color = 'blue')
ax.scatter(x[cluster_id == 1], y[cluster_id == 1], marker = '+', color = 'blue', label  = 'tsne 2')
plt.legend()
plt.title("Two separate runs of t-SNE plotted simultaneously")
plt.show()
```

### Rescale

This one's lazy. Right now I just compute the range of each point cloud $P$ and $Q$ in the x-axis and y-axis, then rescale $P$ to have the same ranges as $Q$. 

`pcu.rescale(P,Q)` returns $P'$ which has the same scale as $Q$.

```python
scaled = pcu.rescale(X_embed_2, X_embed)

fig, ax = plt.subplots() 

x = X_embed[:,0]
y = X_embed[:,1]
ax.scatter(x[cluster_id == 0], y[cluster_id == 0], marker = '^', color = 'red')
ax.scatter(x[cluster_id == 1], y[cluster_id == 1], marker = '+', color = 'red', label = 'tsne 1')

x = scaled[:,0]
y = scaled[:,1]
ax.scatter(x[cluster_id == 0], y[cluster_id == 0], marker = '^', color = 'blue')
ax.scatter(x[cluster_id == 1], y[cluster_id == 1], marker = '+', color = 'blue', label  = 'rescaled tsne 2')
plt.legend()
plt.title("Rescaled runs of t-SNE plotted simultaneously")
plt.show()
```

### Align

I implemented a simple SVD alignment method ([this note](https://igl.ethz.ch/projects/ARAP/svd_rot.pdf) explains it clearly). It finds the optimal Euclidean transformation (orthogonal transformation + translation) to minimize sum of squares distance between the two clouds. 

To use, just call `pcu.align(P,Q)` to return a point cloud $P'$ that is aligned with respect to $Q$. 

```python
aligned = pcu.align(X_embed_2, X_embed)

fig, ax = plt.subplots() 

x = X_embed[:,0]
y = X_embed[:,1]
ax.scatter(x[cluster_id == 0], y[cluster_id == 0], marker = '^', color = 'red')
ax.scatter(x[cluster_id == 1], y[cluster_id == 1], marker = '+', color = 'red', label = 'tsne 1')

x = aligned[:,0]
y = aligned[:,1]
ax.scatter(x[cluster_id == 0], y[cluster_id == 0], marker = '^', color = 'blue')
ax.scatter(x[cluster_id == 1], y[cluster_id == 1], marker = '+', color = 'blue', label  = 'aligned tsne 2')
plt.legend()
plt.title("Aligned runs of t-SNE plotted simultaneously")
plt.show()
```

## Distance 

The third method is more or less a wrapper for the `POT` package (Python Optimal Transport). Given two point clouds $P$ and $Q$, it computes the Earth Mover's Distance between them. EMD depends on two choices 
    1. The mass on each point in the cloud 
    2. The choice of distance in the plane 
    
Right now I only allow uniform masses and Euclidean distance. `POT` supports lots of other stuff that we can include down the road if we like. 

```python
print("Original distance is %f" % pcu.dist(X_embed, X_embed_2))
print("Rescaled distance is %f" % pcu.dist(X_embed, scaled))
print("Aligned distance is %f" % pcu.dist(X_embed, aligned))
```

## Sample Code to get a Null 

With all of this is hand, generating a null is now cleaner. When I show the null in this well-separated case, it still is slightly bimodal. I think this is because the alignment procedure doesn't always work as hoped, but I'm not sure why. 


```python
tot = 100 

null = np.zeros(tot) 

for i in range(tot): 
    # Fit a new t-SNE 
    Y = tsne().fit_transform(X)
    
    # Rescale and align to reference t-SNE 
    Y = pcu.rescale(Y, X_embed)
    Y = pcu.align(Y, X_embed) 
    
    # Compute distance and store 
    null[i] = pcu.dist(Y, X_embed)
```
